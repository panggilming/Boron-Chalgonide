```python
import numpy as np
import matplotlib.pyplot as plt

# Define input data filenames
input_files = ['temp900.dat', 'temp1000.dat', 'temp1100.dat']

# Define the adjusted temperature values for labeling
adjusted_temperatures = [900, 1000, 1100]

# Define colors for the lines
line_colors = ['orange', 'purple', 'red']

# Initialize lists to store maximum and minimum Seebeck coefficients
max_seebeck_values = []
min_seebeck_values = []

# Load and plot data for each temperature
for i, input_file in enumerate(input_files):
    data = np.loadtxt(input_file)
    temperature = adjusted_temperatures[i]  # Adjusted temperature values
    energy_shifted = data[:, 0] * 13.605703976 -  0.4623
    seebeck_coefficient = data[:, 4] / 1000  # Convert from V/K to mV/K
    max_seebeck = np.max(seebeck_coefficient)  # Calculate the maximum Seebeck coefficient
    min_seebeck = np.min(seebeck_coefficient)  # Calculate the minimum Seebeck coefficient
    max_seebeck_values.append(max_seebeck)
    min_seebeck_values.append(min_seebeck)
    plt.plot(energy_shifted, seebeck_coefficient, label=f'{temperature}K', color=line_colors[i])

# Print maximum and minimum Seebeck coefficients
for temperature, max_seebeck, min_seebeck in zip(adjusted_temperatures, max_seebeck_values, min_seebeck_values):
    print(f"Maximum Seebeck coefficient at {temperature}K: {max_seebeck} mV/K")
    print(f"Minimum Seebeck coefficient at {temperature}K: {min_seebeck} mV/K")

# Set plot properties
plt.xlabel(r"${\rm E} - E_{\rm F} \, {\rm (eV)}$", fontsize=14)
plt.ylabel("Seebeck Coefficient (mV/K)", fontsize=14)  # Adjusted y-axis label
plt.legend()

# Set x-axis range
#plt.xlim(-3, 3)

# Put the legend in the upper right corner
plt.legend(prop={'size': 10})  # Adjusted legend size

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define input data filenames
input_files = ['temp200.dat', 'temp300.dat', 'temp400.dat', 'temp500.dat', 'temp600.dat', 'temp700.dat', 'temp800.dat', 'temp900.dat', 'temp1000.dat']

# Define the adjusted temperature values for labeling
adjusted_temperatures = [200, 300, 400, 500, 600, 700, 800, 900, 1000]

# Define colors for the lines
line_colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'orange', 'purple']

# Initialize lists to store maximum Seebeck coefficients
max_seebeck_values = []
min_seebeck_values = []

# Load and plot data for each temperature
for i, input_file in enumerate(input_files):
    data = np.loadtxt(input_file)
    temperature = adjusted_temperatures[i]  # Adjusted temperature values
    energy_shifted = data[:, 0] * 13.605703976 - 2.96
    seebeck_coefficient = data[:, 4] * 1000  # Multiply the y-axis data by 1000 to change from V/K to mV/K
    max_seebeck = np.max(seebeck_coefficient)  # Calculate the maximum Seebeck coefficient
    max_seebeck_values.append(max_seebeck)
    plt.plot(energy_shifted, seebeck_coefficient, label=f'{temperature}K', color=line_colors[i])

# Print maximum Seebeck coefficients
for temperature, max_seebeck in zip(adjusted_temperatures, max_seebeck_values):
    print(f"Maximum Seebeck coefficient at {temperature}K: {max_seebeck} mV/K")

for temperature, min_seebeck in zip(adjusted_temperatures, max_seebeck_values):
    print(f"Minimum Seebeck coefficient at {temperature}K: {max_seebeck} mV/K")    

# Set plot properties
plt.xlabel(r"${\rm E} - E_{\rm F} \, {\rm (eV)}$", fontsize=14)
plt.ylabel("Seebeck Coefficient (m V/K)", fontsize=14)
plt.legend()

# Set x-axis range
plt.xlim(-3.5, 0)

# Put the legend in the upper right corner
plt.legend(prop={'size': 8})

# Show the plot
plt.show()
```
